```python
##implementing k-fold cross validation for MultiNomialNB
from sklearn.model_selection import KFold
import pandas as pd
import numpy as np
from sklearn.naive_bayes import MultinomialNB
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics import accuracy_score, f1_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score

data=pd.read_csv("/content/drive/MyDrive/training_t1_sp.csv")
x=data['text']
y=data['category']

accuracy1=[]
scores1=[]
precision1=[]
recall1=[]

accuracy2=[]
scores2=[]
precision2=[]
recall2=[]
#k-fold cross validation
fold=10
kf = KFold(n_splits=fold, shuffle=True, random_state=42)

#count vectorizer, tfidf with multinomialNB
vectorizer = CountVectorizer()
tfidf=TfidfVectorizer()
model1=MultinomialNB()
model2=MultinomialNB()

for train_i,test_i in kf.split(x):
  x_train=[x[i] for i in train_i]
  y_train=[y[i] for i in train_i]
  x_test=[x[i] for i in test_i]
  y_test=[y[i] for i in test_i]

  train_vect1=vectorizer.fit_transform(x_train)
  test_vect1=vectorizer.transform(x_test)

  train_vect2=tfidf.fit_transform(x_train)
  test_vect2=tfidf.transform(x_test)


  #fitting model
  model1.fit(train_vect1,y_train)
  yhat=model1.predict(test_vect1)

  acc= accuracy_score(y_test,yhat)
  accuracy1.append(acc)

  f_score=f1_score(y_test,yhat,pos_label='hs')
  scores1.append(f_score)

  precision=precision_score(y_test,yhat,pos_label='hs')
  precision1.append(precision)

  recall=recall_score(y_test,yhat,pos_label='hs')
  recall1.append(recall)

  model2.fit(train_vect2,y_train)
  yhat=model2.predict(test_vect2)

  acc=accuracy_score(y_test,yhat)
  accuracy2.append(acc)

  f_score=f1_score(y_test,yhat,pos_label='hs')
  scores2.append(f_score)

  precision=precision_score(y_test,yhat,pos_label='hs')
  precision2.append(precision)

  recall=recall_score(y_test,yhat,pos_label='hs')
  recall2.append(recall)


print("\nAVERAGE ACCURACY count vectorizer: ",np.mean(accuracy1))
print("\nAVERAGE F1 SCORE count vectorier: ",np.mean(scores1))
print("\nAVEARGE PRECISION SCORE count vectorizer: ",np.mean(precision1))
print("\nAVEARGE RECALL SCORE count vectorizer: ",np.mean(recall1))

print("\nAVERAGE ACCURACY tfidf: ",np.mean(accuracy2))
print("\nAVERAGE F1 SCORE tfidf: ",np.mean(scores2))
print("\nAVEARGE PRECISION SCORE tfidf: ",np.mean(precision2))
print("\nAVEARGE RECALL SCORE tfidf: ",np.mean(recall2))

```

```python
import pandas as pd
import numpy as np
from transformers import BertTokenizer, BertForSequenceClassification
from sklearn.model_selection import train_test_split

data=pd.read_csv("/content/drive/MyDrive/training_t1_sp.csv")
x=data['text']
y=data['category']
given_labels=['hs','nhs']

train_x,test_x,train_y,test_y=train_test_split(x,y,train_size=0.7,random_state=42)

#IIC/bert-base-spanish-wwm-cased-ctebmsp
tokenizer = BertTokenizer.from_pretrained('dccuchile/bert-base-spanish-wwm-cased')
model = BertForSequenceClassification.from_pretrained('dccuchile/bert-base-spanish-wwm-cased')


inputs = tokenizer(list(train_x), padding=True, truncation=True, return_tensors="pt")
```

```python
train_x.describe()
```

```python
from sklearn.preprocessing import LabelEncoder
from torch.utils.data import DataLoader
from transformers import AdamW

#labels=[1,0]
encoder=LabelEncoder()
encoder.fit(train_y)
labels=encoder.transform(train_y)

print(labels)

# Prepare DataLoader
dataset = list(zip(inputs["input_ids"], inputs["attention_mask"],labels))
loader = DataLoader(dataset, batch_size=30, shuffle=True)

# Optimizer
optimizer = AdamW(model.parameters(), lr=1e-5)

# Training loop
model.train()
for epoch in range(2):
    print("epoch ",epoch," has started")
    count=0
    for batch in loader:
        count+=1
        print("executing ",count," th time")
        input_ids, attention_mask, labels = batch
        outputs = model(input_ids, attention_mask=attention_mask, labels=labels)
        loss = outputs.loss
        loss.backward()
        optimizer.step()
        optimizer.zero_grad()
    print("epoch completed")
```
